```python
## fonction (celle qui est sur Cloud Firestore)

def fonction_eleve(user_uid):
        #Travail en cours sur le code cloud functions :
    
    import json
    import statistics
    # [END functions_firebase_reactive]

   
    # boucle pour un élève :
    parcoursRef = firestore_db.collection("parcours")
    parcours = parcoursRef.stream()
    liste_moyennes = []
    liste_parcours = []
    liste_evaluations = []
    
 
   
    init_data = {
            u'date': firestore.SERVER_TIMESTAMP,
            u'score': int(0)
        }
    
    init_empty_data = {
        
    }
    
    historique = firestore_db.collection("eleves").document(user_uid).collection("progression").document("historique")
    doc = historique.get()
    if doc.exists:
        print("ok")
    else:
        firestore_db.collection("eleves").document(user_uid).collection("progression").document("historique").set(init_data)
        
        

   
   
    parcoursRef = firestore_db.collection("parcours")
    
    parcours = parcoursRef.stream()
    
    
    for parcour in parcours:
        liste_parcours.append(parcour.id)
        print("on est dans le parcours : ", parcour.id)
        parcoursEleve = firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour.id)
        doc = parcoursEleve.get()
        if doc.exists:
            parcoursData = firestore_db.collection("parcours").document(parcour.id).collection("activites").document("references")
            doc = parcoursData.get()
            for i in doc.to_dict():
                print("\n\nJe vais rechercher l'activite : ", doc.to_dict()[i], " dans le parcours ", parcour.id, "\n")
                ref = firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour.id).collection(doc.to_dict()[i])
                query = ref.order_by("date", direction=firestore.Query.DESCENDING).limit(1)
                results = query.get()
                print("results : ", results)
                if(results == []):
                    print("Le repertoire de l'activite : ", doc.to_dict()[i], "n'existe pas dans le parcours", parcour.id, ". Il faudra créer le répertoire et ajouter un document avec un score de 0")
                    firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour.id).collection(doc.to_dict()[i]).add(init_data)
                for doci in results:
                    print(doci.to_dict())
                    if doc.exists:
                        print("L'activite : ", doc.to_dict()[i], "existe bien dans le répertoire de l'élève dans le parcours : ", parcour.id, " et contient un document. Rien à faire\n\n")
                    
        else:
            print("Le parcours ", parcour.id, " n'existe pas dans le répertoire de l'élève. Il va falloir le créer avec la liste des activités du parcours et des scores = 0")
            firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour.id).set(init_data)
            parcoursData = firestore_db.collection("parcours").document(parcour.id).collection("activites").document("references")
            doc = parcoursData.get()
            for i in doc.to_dict():
                firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour.id).collection(doc.to_dict()[i]).add(init_data)
                    
           
        


    print(liste_parcours)
    try:
        liste_parcours.remove('scores')
    except:
        print("no scores to remove")
    
    try:
        liste_parcours.remove('historique')
    except:
        print("no historique to remove")
        
    #print('liste parcours pour eleve :', liste_eleves[0], liste_parcours)

    dictionary_activites = {}
    dictionary_parcours = []
    for parcour in liste_parcours:
        doc_ref = firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour)
        
        parcoursData = firestore_db.collection("parcours").document(parcour).collection("activites").document("references")
        doce = parcoursData.get()
        liste_activites = []
        for i in doce.to_dict():
            print("i = ", i)
            liste_activites.append(doce.to_dict()[i])
        print("liste activites : ", liste_activites)
        
        liste_scores = []
        dictionary_activites_temp = []
 
        dict_obj = {}
        ref = firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour).collection("evaluation")
        query = ref.order_by("date", direction=firestore.Query.DESCENDING).limit(1)
        results = query.get()
        evaluation = 0
        for doc in results:
                print(f'{doc.id} => {doc.to_dict()}')
                variable = doc.to_dict()
                evaluation = int(variable["score"])

            #print(variable['score'])
        
        
        for activite in liste_activites:
        
            ref = firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour).collection(activite)
            query = ref.order_by("date", direction=firestore.Query.DESCENDING).limit(1)
            results = query.get()
        
            for doc in results:
                #print(f'{doc.id} => {doc.to_dict()}')
                variable = doc.to_dict()

            #print(variable['score'])
            score = variable["score"]
            liste_scores.append(int(score))
        
            print(f"dans le parcours {parcour} pour l\'activite {activite}, score trouve : {score}")
        
            if (score == None):
                score = 0
        
        
            dict_obj.update({activite: score})
            #print("je suis milu :", dict_obj)

        
 

        if liste_scores != []:
            #print('blabla')
            mean = statistics.mean(liste_scores)
        else:
            mean = 0
        
        dict_obj.update({"score": int(mean)})
        dict_obj.update({"evaluation": int(evaluation)})
        liste_moyennes.append(mean)
        
        print('Total : dans le parcours ', parcour, 'les scores trouvés sont :', liste_scores, 'la moyenne est : ', int(mean) )
        dictionary_activites.update({
            parcour: dict_obj,
            })
        #print('big json = ', dictionary_activites)

        data = {
            u'date': firestore.SERVER_TIMESTAMP,
            u'score': int(mean)
        }
    

    

        # Add a new doc in collection 'cities' with ID 'LA'
        firestore_db.collection("eleves").document(user_uid).collection("progression").document(parcour).collection("score").add(data)
        if(parcour == "Starter_Kit"):
            firestore_db.collection("eleves").document(user_uid).collection("progression").document("Starter_Kit").collection("evaluation").add(data)
            
        liste_evaluations.append(evaluation)
    #print("dede", dictionary_activites) 
    # Serializing json
    print("liste eval : ", liste_evaluations)
    print("liste train : ", liste_moyennes)
    
   
    try:
        total_mean = statistics.mean(liste_moyennes)
    except:
        total_mean = 0
    try:
        evaluation_mean = statistics.mean(liste_evaluations)
    except:
        evaluation_mean = 0
    
     
    print("moyenne eval : ", evaluation_mean)
    print("moyenne train :", total_mean)
        
    dictionary_activites.update({"score" : int(total_mean)})
    dictionary_activites.update({"evaluation" : int(evaluation_mean)})
    json_object = json.dumps(dictionary_activites)
    print("\n\nle json est :", json_object)
    
    global_data = {
            u'date': firestore.SERVER_TIMESTAMP,
            u'score': round(evaluation_mean,2)
        }
    
    global_data2 = {
            u'date': firestore.SERVER_TIMESTAMP,
            u'score': round(total_mean,2)
        }
    
    print(global_data2)
    print(global_data)
    

  
    firestore_db.collection("eleves").document(user_uid).collection("progression").document("scores").set(dictionary_activites)
    firestore_db.collection("eleves").document(user_uid).collection("progression").document("historique").collection("scoreEvaluation").add(global_data)
    firestore_db.collection("eleves").document(user_uid).collection("progression").document("historique").collection("scoreEntrainement").add(global_data2)






      
```
